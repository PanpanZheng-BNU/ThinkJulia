```julia
function polygon(t,len,n)
    @svg begin
        for i in 1:n
            forward(t,len)
            turn(t,-360/n)
        end
    end
end
```

```julia
polygon(turtles,2,60)
```

```julia
function arc(t,r,a)
    len = 2*pi*r/360
    @svg begin
        for i in 1:a
            forward(t,len)
            turn(t,-360/360)
        end
    end
end
```

```julia
arc(turtles,100,160)
```

```julia
"""
ployline(t,n,len,angle)

Draws n line segments with the given length and angle (in degrees) between them. t is a turtle. 
"""
function ployline(t,n,len,angle)
    for i in 1:n
        forward(t, len)
        turn(t, -angle)
    end
end
```

```julia
function arc(t,r,a)
    len = 2*pi*r/360
    @svg begin
        for j in 1:7
        for i in 1:a
            forward(t,len)
            turn(t,-360/360)
        end
        turn(t,-180+360/(2*7))
        for i in 1:a
            forward(t,len)
            turn(t,-360/360)
        end
        turn(t,-180+360/(2*7))
        end
    end            
end
```

```julia
turtles=Turtle()
arc(turtles,200,360/7)
```
