## if Condition
### **Chained Conditionals**
Each condition is checked in order. If the first is false, the next is checked, and so on. If one of them is true, the corresponding branch runs and the statements ends. Even if more thean one conditions is true, ***<a style="color:#D0104C">only the first true branch runs.</a>***

```julia
x = 1
if x < -10
    println("x<-10")
elseif x < 0
    println("x<0")
elseif x < 10
    println("x<10")
elseif x < 20
    println("x<20")
end
```

### ~Nested Conditionals~

```julia
x,y = 2,2
if x == y
    println("x is equal to y")
else
    if x > y
        println("x is greater than y")
    else
        println("x is less than y")
    end
end
```

## Recursion

```julia
function countdown(n)
    if n <= 0
        println("Blastoff!")
    else
        println(n)
        countdown(n-1)
    end
end
```

```julia
countdown(10)
```

```julia
function printn(s,n)
    if n <= 0
        return
    end
    println(s)
    printn(s,n-1)
end
```

```julia
printn("hello world",2)
```

```julia
function recurse()
    print("从前有座山，山里有座庙，庙里有个老和尚对小和尚说：")
    recurse()
end
```

```julia
recurse()
```
